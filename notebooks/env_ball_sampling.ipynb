```python
from cobra.io import read_sbml_model

# ======= 加载模型（替换成你的模型路径）=======
model_path = "../models/fixed_models/MAG001.sbml_gapfilled_noO2.xml"
model = read_sbml_model(model_path)
print(f"✅ Loaded model: {model.id}, {len(model.reactions)} reactions, {len(model.metabolites)} metabolites")

# ======= 优化模型 =======
solution = model.optimize()
if solution.status != 'optimal':
    print(f"❌ Optimization failed: {solution.status}")
else:
    print(f"✅ Biomass = {solution.objective_value:.5f}")

# ======= 打印非零内部反应 flux（非 EX_）=======
def print_internal_fluxes(model, solution):
    print("\n📊 非零内部反应 flux（前20条）：")
    count = 0
    for rxn in model.reactions:
        if not rxn.id.startswith("EX_") and abs(solution.fluxes[rxn.id]) > 1e-6:
            print(f"{rxn.id:30} | Flux: {solution.fluxes[rxn.id]:.3f} | {rxn.reaction}")
            count += 1
            if count >= 20:
                break
    if count == 0:
        print("⚠️ 无非零内部反应 flux，可能模型内部无活跃代谢。")

# ======= 打印非零 exchange 反应 flux（EX_）=======
def print_nonzero_exchanges(model, solution):
    print("\n🌍 非零 exchange flux（真实进出物质）:")
    count = 0
    for rxn in model.exchanges:
        flux = solution.fluxes[rxn.id]
        if abs(flux) > 1e-6:
            print(f"{rxn.id:25} | Flux: {flux:.3f} | {rxn.reaction}")
            count += 1
    if count == 0:
        print("⚠️ 所有 exchange flux 都为 0，可能未设置正确的上下限或未记录。")

# ======= 打印潜在 dummy/sink 反应（非 exchange）=======
def print_suspicious_drain_reactions(model):
    print("\n❓ 非标准泄露反应（可能绕开exchange）:")
    dummy_like_rxns = [rxn for rxn in model.reactions
                       if any(tag in rxn.id for tag in ['DM_', 'sink_', 'EX_']) and not rxn.id.startswith("EX_")]
    if dummy_like_rxns:
        for rxn in dummy_like_rxns:
            print(f"{rxn.id:25} | {rxn.reaction}")
    else:
        print("✅ 没有发现非标准的 'DM_'、'sink_'、'EX_' 泄露反应")

# ======= 执行诊断函数 =======
print_internal_fluxes(model, solution)
print_nonzero_exchanges(model, solution)
print_suspicious_drain_reactions(model)
```

```python
import os
import pickle
import pprint
from cobra.io import read_sbml_model

# ========= 读取交换反应 =========
def read_exchange_reactions(tsv_path):
    with open(tsv_path, 'r') as f:
        return [line.strip() for line in f if line.strip()]

EXCHANGE_REACTIONS = read_exchange_reactions(ex_react_path)
print(f"✅ Loaded {len(EXCHANGE_REACTIONS)} exchange reactions")

# ========= 读取模型 =========
model_files = sorted([f for f in os.listdir(models_dir) if f.endswith(".xml")])
models = {}
for filename in model_files[:2]:  # 只加载前两个模型
    model_path = os.path.join(models_dir, filename)
    model_id = os.path.splitext(filename)[0]
    try:
        model = read_sbml_model(model_path)
        models[model_id] = model
        print(f"✅ Loaded model: {model_id}")
    except Exception as e:
        print(f"❌ Failed to load {filename}: {e}")

# ========= 加载环境 =========
with open(env_ball_path, "rb") as f:
    env_ball = pickle.load(f)

env_subset = {k: env_ball[k] for k in list(env_ball.keys())[:2]}  # 只取前两个环境
print(f"✅ Loaded {len(env_subset)} environments")

# ========= 打印一个环境和模型的 exchange 反应（调试用） =========
print("\n🧪 Sample environment input:")
pprint.pprint({k: v for k, v in env_subset[1].items() if v > 0})

print("\n📋 Exchange reactions in first model:")
example_model = list(models.values())[0]
for rxn in example_model.exchanges:
    print(rxn.id)

# ========= 应用环境 =========
def apply_env(env, model, EXCHANGE_REACTIONS):
    for reaction_id in EXCHANGE_REACTIONS:
        if model.reactions.has_id(reaction_id):
            rxn = model.reactions.get_by_id(reaction_id)
            rxn.lower_bound = -1 * env.get(reaction_id, 0.0)  # 摄入：负数
            rxn.upper_bound = 1000  # 显式允许排出

    try:
        solution = model.optimize()
        if solution.status != 'optimal':
            print(f"⚠️ Optimization failed for model {model.id}, status: {solution.status}")
            return None, [0.0] * len(EXCHANGE_REACTIONS)

        fluxes = [
            solution.fluxes[reaction_id] if model.reactions.has_id(reaction_id) else 0.0
            for reaction_id in EXCHANGE_REACTIONS
        ]
        return solution.objective_value, fluxes
    except Exception as e:
        print(f"❌ Optimization error for model {model.id}: {e}")
        return None, [0.0] * len(EXCHANGE_REACTIONS)

# ========= 模拟所有模型和环境 =========
def simulate_all_envs(env_ball, models, exchange_reaction_list):
    result_dict = {}
    for env_idx, env in env_ball.items():
        result_dict[env_idx] = {}
        for model_id, model in models.items():
            print(f"\n🔍 Simulating model {model_id} in environment {env_idx}")
            model_copy = model.copy()
            biomass, fluxes = apply_env(env, model_copy, exchange_reaction_list)
            result_dict[env_idx][model_id] = (biomass, fluxes)
            print(f"✅ Biomass = {biomass}")
            # 调试输出非零交换通量
            nonzero_fluxes = [(rid, flux) for rid, flux in zip(exchange_reaction_list, fluxes) if abs(flux) > 1e-6]
            if nonzero_fluxes:
                print(f"📤 Nonzero exchange fluxes: {nonzero_fluxes[:5]}")
            else:
                print("⚠️ All exchange fluxes are 0")
    return result_dict

# ========= 执行模拟 =========
result = simulate_all_envs(env_subset, models, EXCHANGE_REACTIONS)

# ========= 保存结果 =========
with open("test_simulation_result.pickle", "wb") as f:
    pickle.dump(result, f)

print("\n✅ Test simulation completed and saved.")
```

```python
import os
import xml.etree.ElementTree as ET

folder_path = '../models/Gapfilled_models/'  
xml_files = [f for f in os.listdir(folder_path) if f.endswith('.xml')]

found_files = []  

for xml_file in xml_files:
    file_path = os.path.join(folder_path, xml_file)
    try:

        tree = ET.parse(file_path)
        root = tree.getroot()
        
        xml_string = ET.tostring(root, encoding='utf-8').decode('utf-8')
        
        if 'propionate' in xml_string.lower():  
            found_files.append(xml_file)  
    except Exception as e:
        print(f"read {xml_file} error: {e}")

print("Found files containing 'propionate':")
print(found_files)
```

```python
import cobra
import os

models_with_propionate = []

for name in found_files:
    path = os.path.join("../models/Gapfilled_models", name)
    try:
        model = cobra.io.read_sbml_model(path)
        found = False
        for met in model.metabolites:
            if 'propionate' in met.name.lower():
                print(f"✅ Found in: {name}")
                print(f"  ID: {met.id}, Name: {met.name}, Compartment: {met.compartment}")
                found = True
                break  
        if found:
            models_with_propionate.append(name)
        else:
            print(f" Not found in: {name}")
    except Exception as e:
        print(f" Error reading {name}: {e}")

print("\n Models with propionate:")
print(models_with_propionate)


```
